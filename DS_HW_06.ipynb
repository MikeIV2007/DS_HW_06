# **First dataset: data_2d.csv**

```python
import pandas as pd

data_file_path = "/content/drive/MyDrive/Colab_data/data_hw6/data_2d.csv"
df_01 = pd.read_csv(data_file_path)
display(df_01)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_values = []

k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, init = "k-means++", random_state=42)
    kmeans.fit(df_01)
    inertia_values.append(kmeans.inertia_)

plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()
```

**Кількість кластерів - 2**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

number_of_clusteres = 2
initialization = 'k-means++'

kmeans = KMeans(n_clusters=number_of_clusteres, init = initialization, random_state=42)
kmeans.fit(df_01)

centroids = kmeans.cluster_centers_

colors = plt.cm.tab10(range(number_of_clusteres))

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')

for cluster_label in range(number_of_clusteres):
    ax.scatter(df_01[kmeans.labels_ == cluster_label].iloc[:, 0],
               df_01[kmeans.labels_ == cluster_label].iloc[:, 1],
               df_01[kmeans.labels_ == cluster_label].iloc[:, 2],
               label=f'Cluster {cluster_label + 1}')

for i, centroid in enumerate(centroids):
    ax.scatter(centroid[0], centroid[1], centroid[2], c=[colors[i]], marker='*', s=200)

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title(f'Data with {number_of_clusteres} Clusters and Centroids')
ax.legend()
plt.show()
```

# **Second dataset: mnist.csv**

```python
import pandas as pd

df_02 = pd.read_csv("/content/drive/MyDrive/Colab_data/data_hw6/mnist.csv")
display(df_02)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia_values = []

cluster_range = range(1, 21)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_02)
    inertia_values.append(kmeans.inertia_)

plt.plot(cluster_range, inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

number_of_clusters = 4

pca = PCA(n_components=2, random_state=42)
df_02_2d = pca.fit_transform(df_02)

kmeans = KMeans(n_clusters=number_of_clusters, random_state=42)
kmeans.fit(df_02_2d)
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
for cluster_label in range(number_of_clusters):
    plt.scatter(df_02_2d[cluster_labels == cluster_label, 0],
                df_02_2d[cluster_labels == cluster_label, 1],
                label=f'Cluster {cluster_label + 1}')

for i, centroid in enumerate(centroids):
    cluster_color = plt.cm.tab10(i)
    plt.scatter(centroid[0], centroid[1], c=cluster_color, marker='*', s=200, label=f'Centroid {i+1}')

plt.title(f'K-means Clustering with PCA (2D) and Centroids for {number_of_clusters}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

number_of_clusters = 10

pca = PCA(n_components=2, random_state=42)
df_02_2d = pca.fit_transform(df_02)

kmeans = KMeans(n_clusters=number_of_clusters, random_state=42)
kmeans.fit(df_02_2d)
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

plt.figure(figsize=(8, 6))
for cluster_label in range(number_of_clusters):
    plt.scatter(df_02_2d[cluster_labels == cluster_label, 0],
                df_02_2d[cluster_labels == cluster_label, 1],
                label=f'Cluster {cluster_label + 1}')

for i, centroid in enumerate(centroids):
    cluster_color = plt.cm.tab10(i)
    plt.scatter(centroid[0], centroid[1], c=cluster_color, marker='*', s=200, label=f'Centroid {i+1}')

plt.title(f'K-means Clustering with PCA (2D) and Centroids for {number_of_clusters}')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
